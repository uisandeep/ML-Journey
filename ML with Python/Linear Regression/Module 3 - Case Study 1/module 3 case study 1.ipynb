```python
# Question 2A: New column.

df["total_benefitted"] = df[['No. of Inmates benefitted by Elementary Education', 'No. of Inmates benefitted by Adult Education',
                              'No. of Inmates benefitted by Higher Education', 'No. of Inmates benefitted by Computer Course'
                             ]].sum(axis=1)
df
```

```python
#Question 2B - Create a new row - “totals” that is the sum of all inmates benefitted through each mode across all states.

#df.loc['Column_Total']= df.sum(numeric_only=True, axis=0)
#df
df.loc["total"] = df[['No. of Inmates benefitted by Elementary Education', 'No. of Inmates benefitted by Adult Education',
                              'No. of Inmates benefitted by Higher Education', 'No. of Inmates benefitted by Computer Course'
                             ]].sum(axis=0)
df
```

```python
# Question 3a. Bar plot
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(28,12))
plt.title("Inmates benefitted by different Education Programs")
plt.xlabel("Education Program")
plt.ylabel("Number of Inmates Benefitted")
plt.bar(df['STATE/UT'], df['total_benefitted'])

#Andhra pradesh has maximum number of beneficiares
```

```python
#Question 3b: Make a pie chart that depicts the ratio among different modes of benefits.
plt.figure(figsize=(18,9))
m_dict=dict()
m_dict = { "Elementary Education": df['No. of Inmates benefitted by Elementary Education'].sum(),
           "Adult Education": df['No. of Inmates benefitted by Adult Education'].sum(),
            "Higher Education": df['No. of Inmates benefitted by Higher Education'].sum(),
          "Computer Course": df['No. of Inmates benefitted by Computer Course'].sum()
            
}
plt.pie(m_dict.values(),labels=m_dict.keys())
plt.show()

#We can see that Adult education is more and least is computer courses.
```
