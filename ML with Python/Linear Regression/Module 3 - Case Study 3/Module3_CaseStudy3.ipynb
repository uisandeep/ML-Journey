```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  
%matplotlib inline 
import seaborn as sns
from scipy.stats import spearmanr
from sklearn import linear_model
```

```python
df = pd.read_csv('FyntraCustomerData.csv')
df.head()
```

Question 1: Compute -- Use seaborn to create a jointplot to compare the Time on Website and 
    Yearly Amount Spent columns. Is there a correlation?
 

```python
sns.jointplot(x='Time_on_Website', y='Yearly_Amount_Spent', data=df, stat_func=spearmanr)
```

Question 2: Compute – Do the same as above but now with Time on App and Yearly Amount Spent. 
Is this correlation stronger than 1st One?

```python
sns.jointplot(x='Time_on_App', y='Yearly_Amount_Spent', data=df, stat_func=spearmanr)
```
